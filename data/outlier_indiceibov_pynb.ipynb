**Importando as bibliotecas necessárias**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# **Definindo uma variavel para o outlier do indice do Ibovespa.**

```python
outlier_indiceibov=pd.read_csv('/ibov_indice.csv')
print(outlier_indiceibov)
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=outlier_indiceibov['^BVSP'])
plt.title('Outlier do índice')
plt.show()
```

É possível visualizar que, durante um certo periodo, o valor do indice ^IBOV  esteve abaixo de 70000 — indicando um comportamento diferente do usual.

Portanto, o código a seguir mostra em que periodo isso aconteceu:

```python
outlier_indiceibov[outlier_indiceibov['^BVSP'] < 70000]
```


A queda do indice no periodo **2018-06-18** esta relacionada com a instalibilidade politica brasileira, em razao de ser ano eleitoral e, alem disso, a uma possivel guerra comercial entre EUA e China em 2018.

Os anos seguintes, a partir de 2020, a queda do indice se deu em virtude da Pandemia de Covid-19.
